```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x_train = np.array([50, 55, 60, 65, 70])  # wheight
y_train = np.array([160, 165, 170, 172, 175])  # height

print(f"x train: {x_train}")
print(f"y train: {y_train}")
```

```python
print(f"This is x_train dimensions number: {x_train.shape}")
print(f"This is y_train dimensions number: {y_train.shape}")
```

```python
# m is the number of training examples
m = x_train.shape[0]   # number of elements in the first dimension of x_train
print(f"number of training examples is: {m}")
```

#### Training example x_i, y_i

```python
i = 0

for i in range(len(x_train)):
    print(f"(x^{i}, y^{i}) : ({x_train[i]}, {y_train[i]})")  # training set
```

```python
# Plotting the data points
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
# Setting the title
plt.title("Height Prediction Model")
# Set the y-axis label
# Set the y-axis label
plt.ylabel('Height (cm)')
# Set the x-axis label
plt.xlabel('Weight (kg)')
plt.show()
```
